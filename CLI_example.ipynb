```python
def display(l):
  for i in range(len(l)):
    print(i, end='\t')
  print()
  print('_\t'*len(l))
  for i in range(4):
    for tube in l:
      if i<len(tube):
        print(tube[i] , end='\t')
      else:
        print(' ', end='\t')
    print()
```
