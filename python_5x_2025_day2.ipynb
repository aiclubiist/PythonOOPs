```python
class Tube:
  balls = []
  def __init__(self, balls_para):
    self.balls = balls_para
  def __repr__(self):
    for i in self.balls:
      print(i, end='\t')
    return 'hello'
  def __str__(self):
    for i in self.balls:
      print(i, end='\t')
    return ''
  def user_denifed_method(self):
    for i in self.balls:
      print(i, end='\t')
    return ''
```

```python
class Satellite:
  def __init__(self, norad_id=0):
    self.norad_id = norad_id
    self.tle = ''
  def set_tle(self, tle):
    self.tle = tle
  def __str__(self):
    return "A satellite with norad id: " + str(self.norad_id)
```

```python
InspireSat1 = Satellite(51657)
```

```python
InspireSat1
```

```python
print(InspireSat1)
```

```python
type(InspireSat1.norad_id)
```

```python
class CommunicationSatellite(Satellite):
  def transmit(self):
    print(self.norad_id, 'is Transmitting')
  def receive(self):
    print('Receiving')
```
